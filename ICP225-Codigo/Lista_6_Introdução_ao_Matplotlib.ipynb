# Questao 1

**Gráfico de linha simples**

> **Descrição:**  
Crie um gráfico de linha com os pontos (1, 1), (2, 4), (3, 9), (4, 16).

# Questao 3

**Comparando `plot()` e `scatter()`**

> **Descrição:**  
Utilize os pontos (1,2), (2,4), (3,1), (4,3) para plotar um gráfico com `plot()` e outro com `scatter()`.  
Defina os limites dos eixos com `plt.axis((0, 5, 0, 5))`.  
Explique a diferença visual entre os dois.


````python
import matplotlib.pyplot as plt

# dados
x = [1, 2, 3, 4]
y = [2, 4, 1, 3]

plt.plot(x, y, 'bo-')  # linha azul com bolinhas
plt.title("Usando plot()")
plt.axis((0, 5, 0, 5))
plt.show()

# scatter sem linha
plt.scatter(x, y, color='red')
plt.title("Usando scatter()")
plt.axis((0, 5, 0, 5))

plt.show()
````

# Questao 4

**Múltiplas curvas no mesmo gráfico**

> **Descrição:**  
Desenhe as curvas \( y = x^2 + 2x + 3 \) e \( y = x^2 \) no mesmo gráfico.  
Utilize cores e estilos diferentes e adicione legenda.


````python
import matplotlib.pyplot as plt
import numpy as np

# valores de x
x = np.linspace(-10, 10, 200)  # 200 pontos entre -10 e 10

# funções
y1 = x**2 + 2*x + 3
y2 = x**2

# plotando as curvas
plt.plot(x, y1, label="y = x² + 2x + 3", color="blue", linestyle="-")
plt.plot(x, y2, label="y = x²", color="red", linestyle="--")

# legenda
plt.legend()

# título e rótulos
plt.title("Múltiplas Curvas")
plt.xlabel("x")
plt.ylabel("y")

# mostrar o gráfico
plt.show()
````

# Questao 8

**Criando uma função própria para plotagem**

> **Descrição:**  
Defina uma função para \( f(x) = x^3 + 2x^2 + 5x + 1 \) em Python usando `def`.  
Use `np.linspace()` para gerar valores no intervalo de -3 a 3.  
Plote o gráfico da função com rótulos nos eixos e grade.


````python
import numpy as np
import matplotlib.pyplot as plt

# Definição da função
def f(x):
    return x**3 + 2*x**2 + 5*x + 1

# Gerar valores de x entre -3 e 3
x = np.linspace(-3, 3, 300)

# Calcular f(x)
y = f(x)

# Plotar gráfico
plt.plot(x, y)
plt.title("Gráfico da função f(x) = x³ + 2x² + 5x + 1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
````

# Questao 9

**Anotando o ponto de inflexão de uma função**

> **Descrição:**  
Considere a função \( f(x) = x^3 \).  
Plote a função no intervalo de -5 a 5 e anote o ponto (0, 0) com uma seta, utilizando o comando `plt.annotate()`.  
A anotação deve conter o texto “ponto de inflexão”.  
Adicione título, rótulos nos eixos e grade.


````python
import numpy as np
import matplotlib.pyplot as plt

# Função
def f(x):
    return x**3

# Valores de x
x = np.linspace(-5, 5, 300)
y = f(x)

# Plot da função
plt.plot(x, y, label='f(x) = x³')

# Anotação no ponto de inflexão (0,0)
plt.annotate('ponto de inflexão', xy=(0, 0), xytext=(1, 100), arrowprops=dict(facecolor='black'))

# Título e rótulos
plt.title('Função f(x) = x³')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

plt.show()
````

# Questao 10

**Vendas de discos de vinil ao longo dos anos**

> **Descrição:**  
O arquivo `vinyl_sales.txt` contém dados anuais de vendas de discos de vinil, no formato:
> ```
> 1973,228  
> 1974,204  
> 1975,164  
> ...  
> 2019,0.336120488
> ```
>
> **Tarefa:**  
Utilize o comando do Python para abrir o arquivo e leia linha a linha.  
Para cada linha, separe o ano e o valor das vendas.  
Armazene os anos em uma lista e os valores das vendas em outra.  
Por fim, plote um gráfico de linha para mostrar a evolução das vendas de discos de vinil ao longo dos anos.  
Adicione título, rótulos nos eixos e grade no gráfico.
>
> **Referência do dataset:**  
> https://www.kaggle.com/datasets/imtkaggleteam/40-years-of-music-industry-sales/


````python
import matplotlib.pyplot as plt


years = []
units = []


with open("vinyl_sales.txt", "r") as f:
    for line in f:
      parts = line.split(",")
      year = int(parts[0])
      unit = float(parts[1])
      years.append(year)
      units.append(unit)



plt.plot(years, units)
plt.title("Vendas de Singles de Vinil por Ano")
plt.xlabel("Ano")
plt.ylabel("Unidades Vendidas (Milhões)")
plt.grid(True)
plt.show()
````

# Questao 11

**Vendas de discos de vinil ao longo dos anos (Gráfico de barras)**

> **Descrição:**  
Modifique o exercício anterior para exibir um gráfico de barras mostrando a evolução das vendas de discos de vinil ao longo dos anos.


````python
import matplotlib.pyplot as plt


years = []
units = []


with open("vinyl_sales.txt", "r") as f:
    for line in f:
      parts = line.split(",")
      year = int(parts[0])
      unit = float(parts[1])
      years.append(year)
      units.append(unit)



plt.bar(years, units)
plt.title("Vendas de Singles de Vinil por Ano")
plt.xlabel("Ano")
plt.ylabel("Unidades Vendidas (Milhões)")
plt.grid(True)
plt.show()
````
